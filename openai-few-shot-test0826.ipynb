````python
import requests

endpoint = ""
deployment_name = "gpt4o"
api_key = ""

headers = {
    "Content-Type": "application/json",
    "api-key": api_key
}
# System Messages
# 당신은 사람들이 정보를 찾는 데 도움을 주는 AI 조수입니다.
# 너는 개발자야. python 코드를 주로 사용하는 개발자인데, 파이썬 코드를 작성하는데 도움을 줘.
 
# User
# 랜덤으로 25개의 동물과 이름을 만들어줘

data = {
    "messages": [
        {
            "role": "system",
            "content": [
                {
                "type": "text",
                "text": "당신은 사람들이 정보를 찾는 데 도움을 주는 AI 조수입니다."
                }
            ]
        },
        {
            "role" : "user",
            "content": [
                {
                "type": "text",
                "text": "동물 이름을 만들어줘"
                }
            ]
        },
        {
            "role" : "assistant",
            "content": [
                {
                "type": "text",
                "text": """
                    다음과 같은 Python 코드로 만들 수 있습니다.
                    ```Python
                    # 동물 목록 생성  
                    animals = ["사자", "호랑이", "코끼리", "기린", "원숭이"]  
                    
                    # 동물 목록 출력  
                    print("동물 목록:", animals)  
                    
                    # 동물 목록에 새로운 동물 추가  
                    animals.append("펭귄")  
                    print("업데이트된 동물 목록:", animals)  
                    
                    # 특정 위치에 동물 추가  
                    animals.insert(2, "얼룩말")  
                    print("얼룩말 추가 후 동물 목록:", animals)  
                    
                    # 특정 동물 제거  
                    animals.remove("원숭이")  
                    print("원숭이 제거 후 동물 목록:", animals)  
                    
                    # 마지막 동물 제거  
                    last_animal = animals.pop()  
                    print("제거된 동물:", last_animal)  
                    print("마지막 동물 제거 후 동물 목록:", animals)  
                    
                    # 동물 목록 정렬  
                    animals.sort()  
                    print("정렬된 동물 목록:", animals)  
                    
                    # 동물 목록 역순 정렬  
                    animals.sort(reverse=True)  
                    print("역순 정렬된 동물 목록:", animals)  
                """
                }
            ]
        },
        {
            "role" : "user",
            "content": [
                {
                "type": "text",
                "text": "랜덤으로 10개의 동물 이름을 만들어줘."
                }
            ]
        },
        {
            "role" : "assistant",
            "content": [
                {
                "type": "text",
                "text": """
                    import random  
  
                    # 동물 목록  
                    animal_list = [  
                        "사자", "호랑이", "코끼리", "기린", "원숭이", "펭귄", "얼룩말",  
                        "고양이", "개", "토끼", "늑대", "여우", "곰", "판다", "하마",  
                        "코뿔소", "캥거루", "다람쥐", "고릴라", "치타", "표범", "수달",   
                        "돌고래", "상어", "악어"  
                    ]  
                    
                    # 이름 목록  
                    name_list = [  
                        "민준", "서준", "도윤", "예준", "시우", "하준", "지호", "준우",  
                        "유준", "준서", "윤우", "건우", "도현", "현우", "지후", "우진",  
                        "선우", "서진", "예찬", "승우", "지훈", "승현", "유찬", "시윤", "민재"  
                    ]  
                    
                    # 랜덤으로 25개의 동물과 이름 쌍을 생성  
                    random_animals_and_names = []  
                    for _ in range(25):  
                        animal = random.choice(animal_list)  
                        name = random.choice(name_list)  
                        random_animals_and_names.append((animal, name))  
                    
                    # 결과 출력  
                    for animal, name in random_animals_and_names:  
                        print(f"동물: {animal}, 이름: {name}")  
                """
                }
            ]
        },
        {
            "role": "user",
            "content": [
                {
                "type": "text",
                "text": "랜덤으로 10개의 고양이 이름을 만들어줘"
                }
            ]
        }
    ],
    "temperature": 0.7,
    "top_p": 0.95,
    "max_tokens": 800
}

response = requests.post(f"{endpoint}openai/deployments/{deployment_name}/chat/completions?api-version=2024-02-15-preview",
              headers=headers,
              json=data)

response_json = response.json()
print(response_json['choices'][0]['message']['content'])
````
